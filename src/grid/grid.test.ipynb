## Random point discretization

```python
def generate_random_coords(count: int):
    random_coords = []
    for _ in range(count):
        x = np.random.uniform(0, 100)
        y = np.random.uniform(0, 100)
        random_coords.append(np.array([x, y]))

    return random_coords
```

```python
rand_coords = generate_random_coords(500)

fit = grid.fit_to_grid(rand_coords)
visualize_points(fit)
```
